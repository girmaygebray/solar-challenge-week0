```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

countries = ['benin', 'sierra_leone', 'togo']
data = {c: pd.read_csv(f'data/{c}_clean.csv') for c in countries}

combined = pd.concat([df.assign(country=c) for c, df in data.items()])

sns.boxplot(x='country', y='GHI', data=combined)
plt.title('GHI Comparison Across Countries')

summary = combined.groupby('country')[['GHI','DNI','DHI']].agg(['mean','median','std'])
display(summary)

f_stat, p_val = f_oneway(*(data[c]['GHI'].dropna() for c in countries))
print(f"ANOVA p-value: {p_val:.4f}")
```
